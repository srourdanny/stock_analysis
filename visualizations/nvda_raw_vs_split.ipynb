```python
import pandas as pd
import datetime as dt
import numpy as np

from pathlib import Path
```

```python
# Read in the pre-split data for NVDA
file_path = Path("../data/nvda_daily.csv")
nvda_raw_df = pd.read_csv(file_path, index_col='Unnamed: 0', parse_dates=True).dropna()
nvda_raw_df.head()
```

```python
# Read in the split-adjusted data for NVDA
file_path = Path("../data/nvda_daily_splits.csv")
nvda_split_df = pd.read_csv(file_path, index_col='date', parse_dates=True).dropna()
nvda_raw_df.head()
```

```python
# Plot the raw NVDA close price from Alpha Vantage
nvda_raw_df['4. close'].plot(title="Raw NVDA Close Prices from Alpha Vantage",
                             ylabel="NVDD Price ($)",
                             xlabel="Year",
                             figsize=(10, 5))
```

```python
# Plot the split-adjusted close price
nvda_split_df['close'].plot(title="Split-Adjusted NVDA Close Prices",
                            ylabel="NVDA Price ($)",
                            xlabel="Year",
                            figsize=(10, 5))
```
